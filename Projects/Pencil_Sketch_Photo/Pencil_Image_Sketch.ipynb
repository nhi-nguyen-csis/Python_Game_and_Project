# Convert an Image to Pencil Sketch

Source: https://www.youtube.com/watch?v=X1lDdodZW4A

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('flower.jpg') # load a color image
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert image from BGR to RGB
plt.figure(figsize=(4,4))
plt.imshow(img)
plt.title("Original Image")
plt.show()
```

```python
# convert orginal image to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize=(4,4))
plt.imshow(img_gray, cmap="gray")
plt.title("GrayScale Image")
plt.show()
```

```python
# Inverting the image
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize = (4,4))
plt.imshow(img_invert, cmap="gray")
plt.title("Inverted Image")
plt.axis("off")
plt.show()
```

```python
# Smoothing the image
img_smoothing = cv2.GaussianBlur(img_invert, (61,61), sigmaX=0, sigmaY=0)
plt.figure(figsize = (4,4))
plt.imshow(img_smoothing, cmap="gray")
plt.title("Blur Image")
plt.axis("off")
plt.show()
```

```python
# Converting the image to pencil sketch
final_image = cv2.divide(img_gray, 255-img_smoothing, scale=255)
plt.figure(figsize = (4,4))
plt.imshow(final_image, cmap="gray")
plt.title("Final Sketch Image")
plt.axis("off")
plt.show()
```

# Combine all steps above in the sketch_image( ) function

```python
import cv2
def sketch_image(path, pixel):
    # Load HEIC image using pyheif
    img = cv2.imread(path) 
    grey_filter = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert image from BGR to gray
    invert = cv2.bitwise_not(grey_filter)
    blur = cv2.GaussianBlur(invert, pixel, 0)
    invertedBlur = cv2.bitwise_not(blur)
    sketch_filter = cv2.divide(grey_filter, invertedBlur, scale=256.0)
    # cv2.imwrite("ouput.png", sketch_filter)

    plt.figure(figsize = (10,10))
    plt.imshow(sketch_filter, cmap="gray")
    title = path[:path.index(".")].title()
    plt.title(f"Final Sketch {title}")
    plt.axis("off")
    plt.show()
    export_image(title+"_sketch.png", )

def export_image(file_name, image_data):
    cv2.imwrite(file_name, image_data)

sketch_image('city.jpeg', (171,171)) 
sketch_image('horse.jpeg', (71,71)) 
sketch_image('pic.jpeg', (31,31)) 
```
